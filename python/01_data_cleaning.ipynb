# 01 - Data Cleaning and Exploration
## NorthStar Urban Mobility and Logistics
### Loading all datasets

```python
import pandas as pd

customers    = pd.read_csv('/content/drive/MyDrive/northstar_data/customers.csv')
orders       = pd.read_csv('/content/drive/MyDrive/northstar_data/orders.csv')
deliveries   = pd.read_csv('/content/drive/MyDrive/northstar_data/deliveries.csv')
drivers      = pd.read_csv('/content/drive/MyDrive/northstar_data/drivers.csv')
vehicles     = pd.read_csv('/content/drive/MyDrive/northstar_data/vehicles.csv')
hubs         = pd.read_csv('/content/drive/MyDrive/northstar_data/hubs.csv')
incidents    = pd.read_csv('/content/drive/MyDrive/northstar_data/incidents.csv')
complaints   = pd.read_csv('/content/drive/MyDrive/northstar_data/complaints.csv')
app_events   = pd.read_csv('/content/drive/MyDrive/northstar_data/app_events.csv')

print("All files loaded successfully!")
print("customers rows:", len(customers))
print("orders rows:", len(orders))
print("deliveries rows:", len(deliveries))
print("drivers rows:", len(drivers))
print("vehicles rows:", len(vehicles))
print("hubs rows:", len(hubs))
print("incidents rows:", len(incidents))
print("complaints rows:", len(complaints))
print("app_events rows:", len(app_events))
```

```python
print("=== CUSTOMERS ===")
print(customers.head())
print("\nShape:", customers.shape)
print("\nColumns:", list(customers.columns))
```

```python
all_data = {
    'customers'  : customers,
    'orders'     : orders,
    'deliveries' : deliveries,
    'drivers'    : drivers,
    'vehicles'   : vehicles,
    'hubs'       : hubs,
    'incidents'  : incidents,
    'complaints' : complaints,
    'app_events' : app_events
}

for name, df in all_data.items():
    print(f"\n{'='*40}")
    print(f"FILE: {name}")
    print(f"Rows: {df.shape[0]}  |  Columns: {df.shape[1]}")
    print(f"Columns: {list(df.columns)}")
```

## Checking Data Quality
Before doing any analysis, I want to check if there are missing values
or any messy data that needs fixing.

```python
# check which files have missing values

print("checking missing values in each file...\n")

for name, df in all_data.items():
    missing = df.isnull().sum().sum()  # total missing values in whole file
    print(f"{name}: {missing} missing values")
```

```python
# lets see exactly which columns have missing values

for name, df in all_data.items():
    missing_cols = df.isnull().sum()
    missing_cols = missing_cols[missing_cols > 0]  # only show columns that have missing values

    if len(missing_cols) > 0:
        print(f"\n{name}:")
        print(missing_cols)
    else:
        print(f"\n{name}: looks clean, no missing values")
```

## Fixing Missing Values
Now I will fill in the missing values so the data is ready for analysis.
Different columns need different approaches.

```python
# fixing missing values
# for numbers i will use the average value
# for text columns i will use "Unknown"

# --- customers ---
customers['loyalty_score'] = customers['loyalty_score'].fillna(
    customers['loyalty_score'].mean()
)
# preferred_channel is text so fill with Unknown
customers['preferred_channel'] = customers['preferred_channel'].fillna('Unknown')

print("customers fixed")

# --- orders ---
orders['booking_channel'] = orders['booking_channel'].fillna('Unknown')
print("orders fixed")

# --- deliveries ---
deliveries['delivery_completed_at'] = deliveries['delivery_completed_at'].fillna('Unknown')
# customer_rating is a number so fill with average
deliveries['customer_rating_post_delivery'] = deliveries['customer_rating_post_delivery'].fillna(
    deliveries['customer_rating_post_delivery'].mean()
)
print("deliveries fixed")

# --- drivers ---
drivers['training_score'] = drivers['training_score'].fillna(
    drivers['training_score'].mean()
)
print("drivers fixed")

# --- vehicles ---
vehicles['battery_health_pct'] = vehicles['battery_health_pct'].fillna(
    vehicles['battery_health_pct'].mean()
)
print("vehicles fixed")

# --- incidents ---
# resolved_hours is a number so fill with average
incidents['resolved_hours'] = incidents['resolved_hours'].fillna(
    incidents['resolved_hours'].mean()
)
print("incidents fixed")

# --- complaints ---
# compensation_amount is a number so fill with average
complaints['compensation_amount'] = complaints['compensation_amount'].fillna(
    complaints['compensation_amount'].mean()
)
print("complaints fixed")

# --- app_events ---
# order_id is intentionally missing for some events
# because some app activity has no order linked to it
print("app_events - order_id left as is (some events have no order)")

print("\nall done!")
```

```python
# verify everything is clean now
print("verifying missing values after fixing...\n")

for name, df in all_data.items():
    missing = df.isnull().sum().sum()
    if name == 'app_events':

        missing = df.drop(columns=['order_id']).isnull().sum().sum()
    print(f"{name}: {missing} missing values remaining")
```

## Fixing Inconsistent Categories
Some columns have the same value written in different ways.
For example AIRPORT and airport mean the same thing.
I need to standardise these before analysis.

```python
# let check what unique values exist in zone/category columns

print("=== customers - home_zone ===")
print(customers['home_zone'].unique())

print("\n=== customers - customer_type ===")
print(customers['customer_type'].unique())

print("\n=== customers - account_status ===")
print(customers['account_status'].unique())

print("\n=== orders - pickup_zone ===")
print(orders['pickup_zone'].unique())

print("\n=== orders - dropoff_zone ===")
print(orders['dropoff_zone'].unique())

print("\n=== orders - service_type ===")
print(orders['service_type'].unique())

print("\n=== deliveries - delivery_status ===")
print(deliveries['delivery_status'].unique())

print("\n=== drivers - base_zone ===")
print(drivers['base_zone'].unique())

print("\n=== vehicles - assigned_zone ===")
print(vehicles['assigned_zone'].unique())
```

## Standardising Zone Names
The zone columns have the same places written in many different ways.
For example AIRPORT, Airport and airport all mean the same thing.
I will fix all of these to one standard format.

```python
# creating a function to standardise zone names
# this maps all the messy versions to one clean version

def fix_zone(value):
    if pd.isna(value):
        return value


    v = str(value).strip().lower()

    if v in ['airport', 'airprt']:
        return 'Airport'
    elif v in ['north', 'nrth']:
        return 'North'
    elif v in ['south', 'sth']:
        return 'South'
    elif v in ['central', 'ctr', 'ctrl']:
        return 'Central'
    elif v in ['east']:
        return 'East'
    elif v in ['west']:
        return 'West'
    elif v in ['riverside', 'riversd']:
        return 'Riverside'
    else:
        return value

print("zone fixing function created")
```

```python
# applying the fix to all zone columns
# in customers
customers['home_zone'] = customers['home_zone'].apply(fix_zone)
print("customers home_zone fixed")

# in orders
orders['pickup_zone'] = orders['pickup_zone'].apply(fix_zone)
orders['dropoff_zone'] = orders['dropoff_zone'].apply(fix_zone)
print("orders pickup_zone and dropoff_zone fixed")

# in drivers
drivers['base_zone'] = drivers['base_zone'].apply(fix_zone)
print("drivers base_zone fixed")

# in vehicles
vehicles['assigned_zone'] = vehicles['assigned_zone'].apply(fix_zone)
print("vehicles assigned_zone fixed")

print("\nall zones fixed!")
```

```python
# let verify the zones are clean now

print("=== customers - home_zone after fix ===")
print(sorted(customers['home_zone'].unique()))

print("\n=== orders - pickup_zone after fix ===")
print(sorted(orders['pickup_zone'].unique()))

print("\n=== orders - dropoff_zone after fix ===")
print(sorted(orders['dropoff_zone'].unique()))

print("\n=== drivers - base_zone after fix ===")
print(sorted(drivers['base_zone'].unique()))

print("\n=== vehicles - assigned_zone after fix ===")
print(sorted(vehicles['assigned_zone'].unique()))
```

## Saving Cleaned Data
Now that the data is clean I will save each file as a new CSV.
This way I don't have to clean it again every time I open the notebook.

```python
# saving all cleaned files to google drive
# i am saving them in a new folder called northstar_clean
# so the original files stay untouched

import os

clean_folder = '/content/drive/MyDrive/northstar_clean'
os.makedirs(clean_folder, exist_ok=True)
print("clean folder created")

customers.to_csv(f'{clean_folder}/customers_clean.csv', index=False)
print("customers saved")

orders.to_csv(f'{clean_folder}/orders_clean.csv', index=False)
print("orders saved")

deliveries.to_csv(f'{clean_folder}/deliveries_clean.csv', index=False)
print("deliveries saved")

drivers.to_csv(f'{clean_folder}/drivers_clean.csv', index=False)
print("drivers saved")

vehicles.to_csv(f'{clean_folder}/vehicles_clean.csv', index=False)
print("vehicles saved")

hubs.to_csv(f'{clean_folder}/hubs_clean.csv', index=False)
print("hubs saved")

incidents.to_csv(f'{clean_folder}/incidents_clean.csv', index=False)
print("incidents saved")

complaints.to_csv(f'{clean_folder}/complaints_clean.csv', index=False)
print("complaints saved")

app_events.to_csv(f'{clean_folder}/app_events_clean.csv', index=False)
print("app_events saved")

print("\nall clean files saved successfully!")
```
